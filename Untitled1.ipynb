```python
import matplotlib.pyplot as plt

# Datos simulados para los 6 sprints
sprints = [1, 2, 3, 4, 5, 6]

# Métricas
tasa_exito_envio = [70, 80, 85, 90, 95, 98]
errores_validacion = [15, 12, 10, 7, 5, 2]
verificacion_layout = [7, 8, 8, 9, 9, 10]
verificacion_funcionalidad = [6, 7, 8, 8, 9, 10]
verificacion_compatibilidad = [6, 7, 7, 8, 9, 10]
documentacion_resultados = [5, 6, 7, 8, 9, 10]
satisfaccion_usuario = [6, 7, 8, 8, 9, 9]
eficiencia_energetica = [5, 6, 7, 8, 8, 9]

# Creación de subplots
fig, axs = plt.subplots(4, 2, figsize=(15, 20))

# Gráfico 1: Tasa de Éxito de Envío
axs[0, 0].plot(sprints, tasa_exito_envio, marker='o')
axs[0, 0].set_title('Tasa de Éxito de Envío')
axs[0, 0].set_xlabel('Sprints')
axs[0, 0].set_ylabel('% Éxito')

# Gráfico 2: Número de Errores de Validación
axs[0, 1].plot(sprints, errores_validacion, marker='o', color='r')
axs[0, 1].set_title('Número de Errores de Validación')
axs[0, 1].set_xlabel('Sprints')
axs[0, 1].set_ylabel('Errores')

# Gráfico 3: Verificación de Layout y Diseño
axs[1, 0].plot(sprints, verificacion_layout, marker='o', color='g')
axs[1, 0].set_title('Verificación de Layout y Diseño')
axs[1, 0].set_xlabel('Sprints')
axs[1, 0].set_ylabel('Puntuación (1-10)')

# Gráfico 4: Verificación de Funcionalidad
axs[1, 1].plot(sprints, verificacion_funcionalidad, marker='o', color='purple')
axs[1, 1].set_title('Verificación de Funcionalidad')
axs[1, 1].set_xlabel('Sprints')
axs[1, 1].set_ylabel('Puntuación (1-10)')

# Gráfico 5: Verificación de Compatibilidad
axs[2, 0].plot(sprints, verificacion_compatibilidad, marker='o', color='orange')
axs[2, 0].set_title('Verificación de Compatibilidad')
axs[2, 0].set_xlabel('Sprints')
axs[2, 0].set_ylabel('Puntuación (1-10)')

# Gráfico 6: Documentación de Resultados
axs[2, 1].plot(sprints, documentacion_resultados, marker='o', color='brown')
axs[2, 1].set_title('Documentación de Resultados')
axs[2, 1].set_xlabel('Sprints')
axs[2, 1].set_ylabel('Puntuación (1-10)')

# Gráfico 7: Satisfacción del Usuario
axs[3, 0].plot(sprints, satisfaccion_usuario, marker='o', color='blue')
axs[3, 0].set_title('Satisfacción del Usuario')
axs[3, 0].set_xlabel('Sprints')
axs[3, 0].set_ylabel('Puntuación (1-10)')

# Gráfico 8: Eficiencia Energética (Ahorro γ)
axs[3, 1].plot(sprints, eficiencia_energetica, marker='o', color='cyan')
axs[3, 1].set_title('Eficiencia Energética (Ahorro γ)')
axs[3, 1].set_xlabel('Sprints')
axs[3, 1].set_ylabel('Puntuación (1-10)')

plt.tight_layout()
plt.show()
```
